```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("/kaggle/input/aids-virus-infection-prediction/AIDS_Classification_5000.csv")
df.columns
```

```python
x=df[[ 'time' , 'trt', 'age', 'wtkg', 'hemo', 'homo', 'drugs', 'karnof',
       'oprior', 'z30', 'preanti', 'race', 'gender', 'str2', 'strat',
       'symptom', 'treat', 'offtrt', 'cd40', 'cd420', 'cd80', 'cd820'
       ]]
y=df[['infected']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
x.info()
```

```python
y.info()
```

```python
Classify=LogisticRegression()
Classify.fit(x_train,y_train)
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print("coef_ are:",Classify.coef_)
print("intercept_ are:",Classify.intercept_)
```

```python
y_pred=Classify.predict(x_test)
score=accuracy_score(y_test,y_pred)
print("Score that my model acheive:",score*100)
confusion=confusion_matrix(y_test,y_pred)
print("Confusion metrics:",confusion)
```
