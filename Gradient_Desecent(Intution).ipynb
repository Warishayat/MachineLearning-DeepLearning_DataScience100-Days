```python
#Let apply gradient descent assuing slope is constant and slope value is= 78.35 and lets assume ste staring value of b is is=0
y_pred=((78.35*x)+0).reshape(4)
y_pred
```

```python
m=78.35 #Slope is contant.
b=0
```

```python
loss_slope=- 2 * np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
#Let take the learning rate is 0.1
learning_rate=0.1
step_size=loss_slope * learning_rate
step_size
```

```python
#Now calculate the new intercept value
b=b-step_size
b
```

```python
y_pred1=((78.35*(x)+b))
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color="red",label="ols")
plt.plot(x,y_pred,color="purple",label="old_pred")
plt.plot(x,y_pred1,color="black",label="New")
plt.legend()
plt.show()
```

```python
#Now doing all these mess work within loop or epoch:
m=78.35
b=-100
lr=0.1

epochs=15

for i in range(epochs):
    loss_slope = - 2 * np.sum( y - m*x.ravel() - b)
    b=(b-(lr * loss_slope))

    y_pred= m * x + b

    plt.plot(x,y_pred)
plt.scatter(x,y)
```

```python
class GradientDesecent_Algoritham:
    def __init__(self,lr,epoch):
        self.m=78.35
        self.b=0
        self.lr=lr
        self.epoch=epoch
    def fit(self,x,y):
        #Calculate the b for the fradient desecent.
        for i in range(self.epoch):
            loss_slope= -2 * np.sum(y - m*x.ravel() -b) #ravel is nothing but for 
            self.b=self.b - (self.lr * loss_slope)

            
            print(loss_slope,self.b)
    
```
