```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import Ridge
```

```python
x,y=load_diabetes(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
lr=SGDRegressor(alpha=0.001,penalty='l2',max_iter=500,learning_rate="constant",eta0=0.1)
```

```python
lr.fit(x_train,y_train)
```

```python
p_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print("The r2_score using sgdr:",r2_score(y_test,p_pred))
```

```python
print("Coef_ are:",lr.coef_)
print(lr.intercept_)
```

```python
l2=Ridge(alpha=1.0,max_iter=500,solver="sparse_cg")
```

```python
l2.fit(x_train,y_train)
```

```python
pred=l2.predict(x_test)
```

```python
r2_score(y_test,pred)
```

```python
print("Coef are:",l2.coef_)
```

```python
print("intercept_",l2.intercept_)
```

```python
class RidgeGradient:
    def __init__(self, alpha, epoch, learning_rate):
        self.alpha = alpha
        self.epoch = epoch
        self.learning_rate = learning_rate
        self.weight_ = None
        self.intercept_ = None

    def fit(self, x_train, y_train):
        # Reshape x_train to a 2D array with ncols columns
        ncols = 10  # Adjust this value based on your desired number of columns
        x_train = np.reshape(x_train, (-1, ncols))

        # Add a column of ones for the intercept term
        x_train = np.insert(x_train, 0, 1, axis=1)

        # Initialize weights and intercept
        self.weight_ = np.ones(x_train.shape[1])
        self.intercept_ = 0

        for _ in range(self.epoch):
            # Compute gradient with regularization
            theta_der = np.dot(x_train.T, x_train).dot(self.weight_) - np.dot(x_train.T, y_train[:, np.newaxis]) - self.alpha * self.weight_

            # Update weights
            self.weight_ -= self.learning_rate * theta_der
            self.intercept_ = self.weight_[0]
            self.weight_ = self.weight_[1:]

        print("Intercept (theta_0):", self.intercept_)
        print("Weights (theta_1 to theta_n):", self.weight_)
```
