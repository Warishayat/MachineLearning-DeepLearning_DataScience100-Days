```python
import pandas as pd
import numpy as np 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.datasets import load_diabetes
```

```python
x,y=load_diabetes(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
lr=Ridge(alpha=1.0,solver='cholesky')
```

```python
lr.fit(x_train,y_train)
```

```python
print("coef_ are:",lr.coef_)
print("intercept_ is here:",lr.intercept_)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred=lr.predict(x_test)
r2_score(y_test,y_pred)*100
```

```python
class RideRegression:
    def __init__ (self,alpha=0.1):
        self.coef_ = None
        self.intercept_ =None
        self.alpha=alpha
    def FitTrian(self,x_train,y_train):
        x_train=np.insert(x_train,0,1,axis=1)
        i=np.identity(x_train.shape[1])

        result=np.linalg.inv(np.dot(x_train.T,x_train) + self.alpha * i).dot(x_train.T).dot(y_train)
        self.intercept_ = result[0]
        self.coef_=result[1:]
        print(self.coef_)
        print(self.intercept_)
    def prediction(self,x_test):
        return np.dot(x_test,self.coef_) + self.intercept_
```

```python
r=RideRegression()
r.FitTrian(x_train,y_train)
y=r.prediction(x_test)
r2_score(y_test,y)*100
```

```python
class GradientDecentReg:
    def __init__(self,learning_rate,epoch,alpha=0.001):
        self.alpha=alpha
        self.learning_rate=learning_rate
        self.coef_=None
        self.intercept_=None
        self.epoch=epoch
    def FitData(self,x_train,y_train):
        self.coef_ = np.ones(x_train.shape[1])
        self.intercept_=0
        x_train=np.insert(x_train,0,1,axis=1)
        theta=np.insert(self.coef_,0,self.intercept_)
        for i in range(self.epoch):
            der_weight = np.dot(x_train.T,x_train).dot(theta) -np.dot(x_train.T,y_train) + self.alpha * theta
            theta = theta - (self.learning_rate * der_weight)
        self.intercept_ = theta[0]
        self.coef_=theta[1:]
        print(self.intercept_)
        print(self.coef_)
    def prediction(self,x_test):
        return np.dot(x_test,self.coef_) + self.intercept_
```

```python
r=GradientDecentReg(learning_rate=0.005,epoch=500)
```

```python
r.FitData(x_train,y_train)
```

```python
y=r.prediction(x_test)
```

```python
r2_score(y_test,y)*100
```
