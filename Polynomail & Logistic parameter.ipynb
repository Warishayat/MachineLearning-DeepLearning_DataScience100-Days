```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.datasets import make_classification
from sklearn.metrics import f1_score,accuracy_score
```

```python
x,y=make_classification(n_samples=100,n_features=3,n_informative=1,n_redundant=1,n_classes=2,n_clusters_per_class=1,n_repeated=0,flip_y=0.010)
```

```python
plt.scatter(x[:, 0],x[:, 1] ,c=y,cmap="winter")
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
#noW polynomial faetures
```

```python
poly=PolynomialFeatures(degree=2)
```

```python
x_train_trans=poly.fit_transform(x_train)
x_test_trans=poly.fit_transform(x_test)
```

```python
model.fit(x_train_trans,y_train)
```

```python
acc=model.predict(x_test_trans)
```

```python
accuracy_score(y_test,acc)
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
x,y=load_breast_cancer(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
Model=LogisticRegression(solver="saga",penalty="l1")
```

```python
Model.fit(x_train,y_train)
```

```python
y_pred=Model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
Model=LogisticRegression(solver="saga",penalty="elasticnet",l1_ratio=0.5)
```

```python
Model.fit(x_train,y_train)
```

```python
y_pred=Model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
