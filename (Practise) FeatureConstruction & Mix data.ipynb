```python
df=sns.load_dataset("titanic")
```

```python
df.columns
```

```python
df=df[['survived', 'pclass','sibsp', 'parch', 'fare']]
```

```python
df.sample(30)
```

```python
df["Family"]=df["sibsp"]+df["parch"]+1
```

```python
df.head(5)
```

```python
def FamilyCounter(x):
    if x==1:
        return 0   #mean the person is alone
    elif x>1 and x<=4:
        return 1  #mean the small family in the data.
    else:
        return 2  #mean a big family.
```

```python
df["Family"]=df["Family"].apply(FamilyCounter)
```

```python
df["Family"].value_counts()
```

```python
df.drop(columns=['sibsp', 'parch'],inplace=True)
```

```python
df=pd.read_csv("titanic.csv")
df.head(5)
```

```python
df["number_n"]=pd.to_numeric(df["number"],errors="coerce",downcast="integer")
```

```python
df["Cabin_num"]=df["Cabin"].str.extract('(\d+)')
```

```python
df["Cabin_cat"]=df["Cabin"].str[0]
```
