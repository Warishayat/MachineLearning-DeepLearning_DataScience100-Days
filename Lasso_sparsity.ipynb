```python
#Lasso craeat sparcity because the alpha formed in the neumerator of m so thats why it creat sparcity but in the ridge regression the alpha formed in the denominator
#alpha formed in denominator so thats why ridge regression not creat sparcity.
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
from sklearn.linear_model import Ridge,Lasso,LinearRegression
from sklearn.metrics import r2_score
```

```python
x,y=load_diabetes(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rid=Ridge(alpha=1.0)
```

```python
rid.fit(x_train,y_train)
```

```python
rid.coef_
```

```python
rid.intercept_
```

```python
y_pred=rid.predict(x_test)
r2_score(y_test,y_pred)*100
```

```python
laso=Lasso(alpha=0.1)
```

```python
laso.fit(x_train,y_train)
```

```python
laso.coef_
```

```python
laso.intercept_
```

```python
result=laso.predict(x_test)
r2_score(y_test,result)*100
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y=lr.predict(x_test)
```

```python
r2_score(y_test,y)*100
```
