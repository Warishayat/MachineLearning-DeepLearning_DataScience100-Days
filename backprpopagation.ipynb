```python
import numpy as np
import pandas as pd
import tensorflow 
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
df=pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]],columns=["cgpa","profile_score","lpa"])
```

```python
df
```

```python
model=Sequential()

model.add(Dense(2,activation="linear",input_dim=2))
model.add(Dense(1,activation="linear"))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
new_weights=[np.array([[0.1 , 0.1],
                       [0.1 , 0.1]],dtype=np.float32),
                     np.array([0. , 0.],dtype=np.float32),
                     np.array([[0.1],
                        [0.1]],dtype=np.float32),
                    np.array([0.],dtype=np.float32)]
```

```python
new_weights
```

```python
model.set_weights(new_weights)
```

```python
model.get_weights()
```

```python
optimizres=keras.optimizers.Adam(learning_rate=0.001)
```

```python
model.compile(loss="mean_squared_error",optimizer=optimizres)
```

```python
#extracat x and y from the data
df
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1:]
```

```python
model.fit(x,y,epochs=75,verbose=1,batch_size=1)
```

```python
model.get_weights()
```

```python
df=pd.DataFrame([[8,8,1],[7,9,1],[6,10,0],[5,5,0]],columns=["cgpa","profile_score","placment"])
```

```python
df
```

```python
#again extracting the data x and y from the data.
x=df.iloc[:,:-1]
y=df.iloc[:,-1:]
```

```python
model=Sequential()
model.add(Dense(2,activation="sigmoid",input_dim=2))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
new_weights=[np.array([[0.1 , 0.1],
                       [0.1 , 0.1]],dtype=np.float32),
                     np.array([0. , 0.],dtype=np.float32),
                     np.array([[0.1],
                        [0.1]],dtype=np.float32),
                    np.array([0.],dtype=np.float32)]
```

```python
model.set_weights(new_weights)
```

```python
model.get_weights()
```

```python
optimizres=keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss="binary_crossentropy",optimizer=optimizres)
```

```python
model.fit(x,y,epochs=50,verbose=1,batch_size=1)
```

```python
model.get_weights()
```
