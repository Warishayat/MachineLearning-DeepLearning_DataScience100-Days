```python
class MultipleRegression:
    def __init__(self):
        self.coef_=None
        self.intercept_=None
    def Fit_Data(self,x_train,y_train):
        #First i will set the columns then apply the formoula which is B=(xt-x)-1 * xt*y
        x_train=np.insert(x_train,0,1,axis=1)
        #calculate now with this formoula which is B=(xt-x)-1 * xt*y
        Betas=np.linalg.inv(np.dot(x_train.T,x_train)).dot(x_train.T).dot(y_train)
        self.intercept_ = Betas[0]
        self.coef_= Betas[1:]
        print(self.intercept_)
        print(self.coef_)
    def prediction(self,x_test):
        y_pred=np.dot(x_test,self.coef_)+self.intercept_
        return y_pred
```

```python
reg=MultipleRegression()
reg.Fit_Data(x_train,y_train)
```

```python
y_pred=reg.prediction(x_test)
print(y_pred)
```
