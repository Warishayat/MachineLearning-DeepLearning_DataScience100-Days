```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.DataFrame({'text':["people watch campux","campux watch campux","people write comment","campux write comment"],'output':[1,1,0,0]})
```

# Aplly BAG OF WORDS TECHNIQUE

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
bow=cv.fit_transform(df["text"])
```

```python
print(cv.vocabulary_)
print(bow.toarray())
```

```python
cv.transform(["campux watch campux and watch campux also"]).toarray()
```

# Applaying N-Gram TECHNIQUE

```python
df=pd.DataFrame({'text':["people watch campux","campux watch campux","people write comment","campux write comment"],'output':[1,1,0,0]})
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(ngram_range=(2,2))
bow=cv.fit_transform(df["text"])
```

```python
print(cv.vocabulary_)
print(bow.toarray())
```

# TF-IDF ( term frequnecy - inverse documnet frequenct )

```python
df=pd.DataFrame({'text':["people watch campux","campux watch campux","people write comment","campux write comment"],'output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_idf = TfidfVectorizer()
tf_idf.fit_transform(df["text"]).toarray()
```

```python
print(tf_idf.idf_) 
print(tf_idf.get_feature_names_out()) 
```
