```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
x = np.array(7 * np.random.rand(100, 1) - 3) 
x1 = x.reshape(-1, 1) 
y = 13 * x*x + 2 * x + 7 
```

```python
plt.scatter(x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred=lr.predict(x_test)
r2_score(y_test,y_pred) #that gave us pathetic result on the linear data
```

```python
plt.plot(x_train,lr.predict(x_train),color="red")
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#Now Applying polynomial on linear Regression
```

```python
poly=PolynomialFeatures(degree=2)
x_train_trans=poly.fit_transform(x_train)
x_test_trans=poly.fit_transform(x_test)
```

```python
print("x_train:",x_train[0])
print("x_train_trans:",x_train_trans[0])
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train_trans,y_train)
```

```python
y_pred=lr.predict(x_test_trans)
```

```python
r2_score(y_test,y_pred)
```

```python
plt.plot(x_train,lr.predict(x_train_trans),color="red")
plt.scatter(x,y,color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
