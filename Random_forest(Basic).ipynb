```python
import pandas as pd
import numpy as np 
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=5,n_informative=2)
```

```python
df=pd.DataFrame(x)
df=df.rename(columns={
    0:"col1",
    1:"col2",
    3:"col3",
    4:"col4",
    5:"col5"}
)
dt=pd.DataFrame(y)
dt=dt.rename(columns={
    0:"target"
})
```

```python
dt.sample(2)
```

```python
df=pd.concat((df,dt),axis=1)
```

```python
import random
```

```python
def row_sample(df,row_percent):
    return df.sample(int(row_percent * df.shape[0]),replace=True)
```

```python
def feature_sample(df,col_percent):
    cols=random.sample(df.columns.tolist()[:-1],int(col_percent * df.shape[1]))
    return df[cols]
```
