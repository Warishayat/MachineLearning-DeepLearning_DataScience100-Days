```python
class LinearRegression:
    def __init__(self):
        self.m=None
        self.b=None
    def Fitdata(self,x_train,y_train):
        neu=0
        den=0
        #first calculate the coefficent
        for i in range(x_train.shape[0]):
            neu += (x_train[i] - x_train.mean()) * (y_train[i] - y_train.mean())
            den += (x_train[i] - x_train.mean()) * (x_train[i] - x_train.mean()) 

        self.m = neu/den
        #Now the intercept is:
        self.b = y_train.mean() - (self.m * x_train.mean())
        print("Intercept_:",self.b)
        print("coef_",self.m)
    def predict_Result(self,x_test):
        return (self.m * x_test) + self.b
```

```python
lr=LinearRegression()
lr.Fitdata(x_train,y_train)
cgpa=int(input("Enter cgpa between 4 to 9:"))
lr.predict_Result([cgpa])
```
