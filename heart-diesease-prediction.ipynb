```python
df=pd.read_csv("/kaggle/input/heart-disease-classification/heart.csv")
df.head(2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score,recall_score,f1_score,classification_report
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1:]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
Model=LogisticRegression()
```

```python
Model.fit(x_train,y_train)
```

```python
result=Model.predict(x_test)
```

```python
#the percentage of actual positives that a model correctly identifies called recall score.
#
print("Precise score is:",precision_score(y_test,result)*100)
print("Recall_Score is:",recall_score(y_test,result)*100)
print("f1 score is:",f1_score(y_test,result)*100)
print("Classification report is:")
print(classification_report(y_test,result))
```
