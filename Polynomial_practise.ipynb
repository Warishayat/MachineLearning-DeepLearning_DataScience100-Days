```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
np.random.seed(0)
x = np.linspace(-5, 5, num=50)
y = 2.0 + 1.5 * x + 3.0 * x**2 + np.random.normal(scale=3.0, size=x.shape)
```

```python
plt.scatter(x,y)
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.plot(x_train,lr.predict(x_train),color="red")
plt.scatter(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
r2_score(y_test,y_pred)*100
```

```python
#Now apply polynomialfeature
```

```python
poly=PolynomialFeatures(degree=2)
x_train_trans=poly.fit_transform(x_train)
x_test_trans=poly.fit_transform(x_test)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train_trans,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.plot(x_train, lr.predict(x_train_trans), color="red")
plt.scatter(x_train, y_train)
plt.show() 
```

```python
#Now check the accuracy
```

```python
test=lr.predict(x_test_trans)
```

```python
r2_score(y_test,test)*100 #No doubt its very good.
```
